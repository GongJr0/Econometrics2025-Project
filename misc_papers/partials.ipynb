```python
import sympy as sp
```

```python
u = sp.symbols(r'u')
u_ss = sp.symbols(r'u^*')
u_gap = sp.symbols(r'\bar{u}')
P_t = sp.symbols(r'P_t')
P_exp = sp.symbols(r'\mathbb{E}_t[P_{t+1}]')
alpha = sp.symbols(r'\alpha')
eps = sp.symbols(r'\varepsilon_t')

exp_inf = (P_exp - P_t) / P_t
```

```python
philips = exp_inf - alpha*(u-u_ss)
philips_gap = exp_inf - alpha*u_gap
log_philips = sp.log(philips)
sp.diff(sp.log(philips_gap), u_gap).simplify()
```

```python
beta = sp.symbols(r'\beta')
C = sp.symbols(r'C')
ar_expectation = philips_gap.subs(P_exp, beta*P_t+C)
log_ar_exp = sp.log(ar_expectation)
sp.diff(log_ar_exp, P_t).simplify()
```

```python
Y = sp.symbols(r'Y')
Y_ss = sp.symbols(r'Y^*')
gamma = sp.symbols(r'\gamma')
output_gap = (Y - Y_ss) / Y_ss
okun = -gamma * (u - u_ss)

sp.diff(okun, u).simplify()
```

```python
log_okun = sp.log(okun)
sp.diff(log_okun, u).simplify()
```
